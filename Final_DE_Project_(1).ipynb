### Oversampling using SMOTE

In the previous classification model, we applied a train test split. If we check the number of instances in the testing data it is found to be 97125. One important thing to keep in mind is that we shouldn't oversample our testing data. In order to do that, we have already taken a sample of 97125 instances fromt the dataset and stored it in mydata_test.

```python
mydata_test = mydata1.sample( n = 97125 )
mydata_train = mydata1[~mydata1.isin(mydata_test)].dropna()
```

To get a dataset not containing the testing data, we get the data which is not in mydata_test, drop the null values and store it in mydata_train.
The isin() method, check whether each row/instance of the dataframes match or not. If they don't match then we choose them and delete the values which matched and became null.

```python
from imblearn.over_sampling import SMOTE

x = mydata_train.drop('Churn', axis=1)
y = mydata_train['Churn']

smote = SMOTE(sampling_strategy='auto')
x, y = smote.fit_resample(x, y)

mydata_train_os = pd.concat([pd.DataFrame(x, x.columns), pd.Series(y, name='Churn')], axis=1)
```

Let's try to understand the above code step by step:



1.   Importing SMOTE from imblearn library
2.   Dividing the data. x is a dataframe containg all the features of mydata_train except the Churn. y is a series containing all the instances of Churn.


3.   Applying SMOTE and setting the sampling_strategy as auto.  This means, that my model will itself try to rationallize the minority (1) and majority class (0). It will decrease the number of instances from the majority class and increase the number of instances in the minority class. We have sampling_strategy as auto because, we won't face the issue of generating synthetic data for the rest of the features of the increased instances of the minority class. If we would have set the sampling_strategy as 1, our model will increase the instances of minorty class to make its size equal to the majority class. This could lead to very large number of rows, and furthermore, we would have had to generate synthetic data for each feature for the increased instances.

2.   After applying the sampling technique, we are storing them in x and y. The last step concatenates the x (dataframe) with y (series) and stores them in mydata_train_os dataframe.




```python
os= mydata_train_os['Churn'].value_counts()
print(os)
os.plot(kind = 'bar')
```

From the above bar chart, we can see that our Churn is now perfectly balanced. It looks like SMOTE decreased the size of Churn = 0 to from 161874 to 84392 and increased the size of churn = 1 from 40744 to 84392.

### Undersampling using Sk learn's resample

```python
from sklearn.utils import resample

mj_class = mydata1[mydata1.Churn==0]
mn_class = mydata1[mydata1.Churn==1]

mj_class_us = resample(mj_class, n_samples = len(mn_class), random_state = 27)

mydata_us = pd.concat([mj_class_us, mn_class])
```

For undersampling, we are trying to decrease size of our majority class (mj_class) to the size of our minority class (mn_class). We have done that using resample() from sklearn library.

```python
us= mydata_us['Churn'].value_counts()
print(us)
us.plot(kind = 'bar')
```

From the above bar chart, we can see that our Churn is now perfectly undersampled. The size of Churn = 0 decreased to from 161874 to 40744 which is equal to the size of the minorty class (churn = 1).
